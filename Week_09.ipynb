# Generators and Iterators

```python
# creating a basic iterator from an iterable
sports = ["baseball", "soccer", "football", "hockey", "basketball"]
my_iter = iter(sports)
print(next(my_iter)) #output first item
print(next(my_iter)) #output second item
for item in my_iter:
    print(item)
#print(next(my_iter)) #will produce error
```

```python
# creating our own iterator
class Alphabet():
    def __iter__(self):
        self.letters = "abcdefghijklmnopqrstuvwxyz"
        self.index = 0
        return self
    def __next__(self):
        if self.index <=25:
            char = self.letters[self.index]
            self.index += 1
            return char
        else:
            raise StopIteration
for char in Alphabet():
    print(char)
```

```python
# creating our own range generator with start, stop, and step parameters
def myRange(stop, start=0, step=1):
    while start < stop:
        print("Generator Start Value: {}".format(start))
        yield start
        start += step # increment start, otherwise infinite loop
for x in myRange(5):
    print("For Loop X Value: {}".format(x))
```

```python
# Reverse Iteration
class RevIter():
    def __init__(self, lista):
        self.lista = lista
        self.index = len(lista) - 1
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.index >= 0:
            valor = self.lista[self.index]
            self.index -= 1
            return valor
        else:
            raise StopIteration
            
for i in RevIter([1, 2, 3, 4, 5]):
    print(i)
```

```python
#Squares:
def myRangeSquare(stop, start=0, step=1):
    while start < stop:
        print(start ** 2)
        yield start
        start += step # increment start, otherwise infinite loop
for x in myRangeSquare(5):
    print("For Loop X Value: {}".format(x + 1))
```

# Tuesday: Decorators

```python
# creating and applying own decorator using the @ symbol

def decorator(func):
    def wrap():
        print("======")
        func()
        print("======")
    return wrap

@decorator
def printName():
    print("John!")
    
printName()
```

```python
# creating a decorator that takes in parameters

def run_times(num):
    def wrap(func):
        for i in range(num):
            func()
    return wrap

@run_times(4)
def sayHello():
    print("Hello!")
```

```python
# creating a decorator for a function that accepts parameters

def birthday(func):
    def wrap(name, age):
        func(name, age + 1)
    return wrap

@birthday
def celebrate(name, age):
    print("Happy birthday {}, you are now {}.".format(name, age))

celebrate("Natan", 32)
```

```python
# real world sim, restricting function access

def login_required(func):
    def wrap(user):
        password = input("What is the password? ")
        if password == user["password"]:
            func(user)
        else:
            print("Access Denied")
    return wrap

@login_required
def restrictedFunc(user):
    print("Access granted, welcome {}".format(user["name"]))
    
user = {"name" : "Jess", "password" : "ilywpf"}

restrictedFunc(user)
```
